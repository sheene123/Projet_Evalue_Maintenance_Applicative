**<h1>Projet pratique : Mini-Convertisseur de devises Streamlit</h1>**

**Objectif du projet**
Chaque groupe de trois étudiants prend une application simple existante, volontairement imparfaite, et applique **toutes les dimensions de la maintenance applicative** :

* **Maintenance corrective** : corriger les bugs existants.
* **Maintenance évolutive** : ajouter des fonctionnalités demandées.
* **Maintenance adaptative** : adapter l’application à de nouvelles contraintes techniques (ex. API externe).
* **Maintenance perfective** : améliorer la lisibilité, la structure et la qualité du code (refactoring et UX).

Le but est de combiner analyse, tests et bonnes pratiques de développement dans un projet concret.

---

**<h2>1. L’application initiale</h2>**

Fichier : `app.py`

```python
import streamlit as st

st.title("Convertisseur de devises")

rates = {
    "EUR": 1,
    "USD": 1.1,
    "JPY": 130
}

amount = st.number_input("Montant :", min_value=0.0, format="%.2f")
from_currency = st.selectbox("De :", rates.keys())
to_currency = st.selectbox("Vers :", rates.keys())

if st.button("Convertir"):
    result = amount * rates[to_currency] / rates[from_currency]
    st.success(f"{amount} {from_currency} = {result:.2f} {to_currency}")
```

**Points à observer avant toute modification :**

* Les taux sont codés en dur.
* La logique de conversion est directement dans l’interface Streamlit.
* Aucune gestion d’erreurs si le montant est nul ou si l’utilisateur choisit la même devise.
* Aucun historique des conversions n’est conservé.

**<h3>D. Maintenance perfective</h3>**

1. Refactoriser la logique métier dans un fichier `app_functions.py`.
2. Créer une fonction `convert(amount, from_currency, to_currency, rates)` qui renvoie le résultat.
3. Créer des tests unitaires dans `test_app.py` pour valider la conversion.
4. Vérifier la qualité du code avec `flake8` et `black`.
5. Mettre en place un workflow GitHub Actions pour exécuter automatiquement les tests à chaque Pull Request.

Exemple de fonction refactorisée :

```python
# app_functions.py
def convert(amount, from_currency, to_currency, rates):
    if amount <= 0:
        return None
    if from_currency == to_currency:
        return amount
    return amount * rates[to_currency] / rates[from_currency]
```

Exemple de test unitaire :

```python
# test_app.py
from app_functions import convert

def test_convert_eur_usd():
    rates = {"EUR": 1, "USD": 1.1}
    assert round(convert(10, "EUR", "USD", rates), 2) == 11.0
```
